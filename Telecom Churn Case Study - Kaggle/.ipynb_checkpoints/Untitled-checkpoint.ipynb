## Cleaning The Data

```python
df.shape
```

```python
len(df.columns[df.isnull().any()])
```

```python
mask = df.isnull().sum()/df.shape[0] *100
mask.loc[mask > 50]
```

```python
## above is the columns that have more than 50% of its values missing
## some of the columns such are important for our analysis so we will go aheat and keep them by filling them with 0
list_to_keep = ['av_rech_amt_data', 'arpu_2g', 'arpu_3g', 'count_rech_2g', 'count_rech_3g','max_rech_data', 'total_rech_data','fb_user','night_pck_user']

for i in [a+b for b in ['_6','_7','_8','_9'] for a in list_to_keep]:
    df[i].fillna(0, inplace=True)
        
        
```

```python
mask2 = df.isnull().sum()/df.shape[0] *100
mask2.loc[mask2 > 50]
```

```python
## we can see that most of the values are filled except for 4
## well go ahead and drop theose 4 from the set
index_to_drop =mask2.loc[mask2 > 50].index
```

```python
index_to_drop = list(index_to_drop)
```

```python
df.drop(index_to_drop, axis = 1, inplace=True)
```

```python
df.shape
```

```python
mask3 = df.isnull().sum()/df.shape[0] *100
len(mask3.loc[mask3 > 5])
```

```python
## there are 58 features having 5% or more missing values
```

```python
missing_vals_ind = list(mask3.loc[mask3 > 5].index)
df[df[missing_vals_ind].isnull().all(axis=1)][missing_vals_ind].head()
```

```python
## above are all the columns that have more than 5% of its data missing, well go ahead and fill them with 0
```

```python
df[missing_vals_ind] = df[missing_vals_ind].fillna(0)
```
