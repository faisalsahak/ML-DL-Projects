```python
# well replace seasons numbers with appropriate season
df['season'] = df['season'].map({1:"spring",2:"summer",3:"fall",4:"winter"})
```

```python
df['weathersit'] = df['weathersit'].map({1: "Clear, Few clouds, Partly cloudy, Partly cloudy", 
                          2: 'Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist',
                          3:'Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds',
                          4:'Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog'})
```

```python
df['mnth'] = df['mnth'].map({1:'Jan',2:'Feb',3:'Mar',4:'Apr',5:'May',6:'Jun',7:'Jul',8:'Aug',9:'Sep',10:'Oct', 11:'Nov',12:'Dec'})
```

```python
df['weekday'] = df['weekday'].map({5:'Sun',6:'Mon',0:'Tue',1:'Wed',2:'Thu',3:'Fri',4:'Sat'})
```

```python



## data preparation

```

```python
#create dummy variables
temp = pd.get_dummies(df[['season', 'weathersit','mnth','weekday']],drop_first=True)
```

```python
temp.head()
```

```python
df = pd.concat([df, temp], axis=1)
```

```python
df.head()
```

```python
df.drop(['season','instant','casual','weekday', 'weathersit','dteday','mnth','registered'], axis = 1,inplace = True)
```

```python
df.columns
```

```python
df.shape
```

```python



#### Train test split the Data



```

```python
df_train, df_test = train_test_split(df, train_size = 0.7, test_size = 0.3, random_state = 100)
print(df_train)
print(df_test)
```

```python
print(df_train.shape)
print(df_test.shape)
```

```python
### Rescale the data using MinMaxScaler

scaler = MinMaxScaler()
var = ['temp', 'atemp', 'hum', 'windspeed', 'cnt']
df_train[var] = scaler.fit_transform(df_train[var])


```

```python
df_train.head()
```

```python
df_train.describe().transpose()
```

```python
# check to see if there are any twho or more variables that have a high correlation
plt.figure(figsize=(25,15))
sns.heatmap(df_train.corr(), annot=True)
```

```python
## split data for x y to build our model

X_train = df_train
y_train = df_train.pop('cnt')
```

```python
X_train_lm = sm.add_constant(X_train)
lrm1 = sm.OLS(y_train,X_train_lm).fit()
```

```python
lrm1.summary()
```

```python



### building model






```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)

rfe = RFE(lr, 15)
rfe = rfe.fit(X_train, y_train)

col = X_train.columns[rfe.support_]
X_train.columns[~rfe.support_]
X_train_rfe = X_train[col]
X_train_rfe = sm.add_constant(X_train_rfe)
lm = sm.OLS(y_train,X_train_rfe).fit()
lm.summary()
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python



#calculating the VIF



```

```python
vif = pd.DataFrame()
X = X_train_rfe
vif['Features']=X.columns
vif['VIF']=[variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF']=round(vif['VIF'], 2)
vif=vif.sort_values(by="VIF", ascending = False)
vif
```

```python
#drop the column with high VIF and p values

newX = X_train_rfe.drop(["mnth_Nov"], axis =1)
# X_train_rfe
```

```python
# Removed one of the variables to see if the model performs better, now run the model again with constant
X_train_lm = sm.add_constant(newX)
lrm2 = sm.OLS(y_train,X_train_lm).fit()
lrm2.summary()
```

```python
# now well remove another variable and see how the model reacts or performs
newX2 = X_train_rfe.drop(["weekday_Mon"], axis =1)
X_train_lm2 = sm.add_constant(newX2)
lrm3 = sm.OLS(y_train,X_train_lm2).fit()
lrm3.summary()
```

```python
vif = pd.DataFrame()
X = newX2
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

```python
#slightly better but not much, well go ahead and do the same with other variables
newX3 = X_train_rfe.drop(["weekday_Mon"], axis =1)
X_train_lm3 = sm.add_constant(newX3)
lrm4 = sm.OLS(y_train,X_train_lm3).fit()
lrm4.summary()
```

```python
vif = pd.DataFrame()
X = newX3
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

```python
newX4 = X_train_rfe.drop(["mnth_Sep"], axis =1)
X_train_lm4 = sm.add_constant(newX4)
lrm5 = sm.OLS(y_train,X_train_lm4).fit()
lrm5.summary()
```

```python
vif = pd.DataFrame()
X = newX4
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

```python
newX5 = X_train_rfe.drop(["mnth_Sep"], axis =1)
X_train_lm5 = sm.add_constant(newX5)
lrm6 = sm.OLS(y_train,X_train_lm5).fit()
lrm6.summary()
```

```python
vif = pd.DataFrame()
X = newX5
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

```python



# Check the Residual of the train data

```

```python
y_train_r2e= lrm6.predict(X_train_lm4)
```

```python
fig = plt.figure(figsize=(8,8))
sns.distplot((y_train - y_train_r2e), bins = 30)
fig.suptitle('Residual Error Dist.', fontsize = 15)
plt.xlabel('Error', fontsize = 15)
```
